```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
# Generate random samples

samples = []

for i in range(10000):
    experiments = np.random.randint(1, 10, size=10000)
    sample_mean = np.mean(experiments)
    samples.append(sample_mean)

# Plot the histogram of the samples
plt.hist(samples, bins=30, density=False, alpha=0.7, edgecolor='white', linewidth=1.5)

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Count')
plt.title('Distribution of Random Variable')

# Display the plot
plt.show()
```
