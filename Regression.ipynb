```python
import random
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = list(range(0,10,1))
```

```python
y = [random.uniform(0, 10) for _ in range(10)]
```

```python
plt.plot(x,y,marker='o')
```

```python
def err(a,b,x,y):
    e = 0
    for i in range(len(x)):
        e = abs( y[i] - (a*x[i] + b) )
    return e
```

```python
a_rng = list(range(-5,5))
```

```python
b_rng = list(range(10))
```

```python
e = []
for i,j in zip(a_rng,b_rng):
    e.append(err(i, j, x, y))
```

```python
e
```

```python
plt.plot(x,y,e)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```
