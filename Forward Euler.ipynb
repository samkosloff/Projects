## Objective

Suppose I have a function $y = f(x)$ but all I know about it is (1) the value of the function at $x = 0$ and (2) the slope of the function $\frac{dy}{dx}$ at each point. How can I recover an estimate of the original function? We can do this using the forward Euler method.

Before we can estimate $f(x)$ we need to choose a step size to increment $x$. This will determine the accuracy of our estimation. Let $h$ be the step size such that $x_n = x_{n-1} + h$. The approach will proceed as follows:

1. Let $x_0,y_0$ be the initial condition and $\frac{dy}{dx}$ be the slope.
2. Therefore $\frac{dy}{dx}(x_0)$ is the slope of $f(x)$ at the point $x_0$.
3. Now $x_n = x_{n-1} + h$ and $\frac{dy}{dx}(x_n)\approx\frac{y_{n+1} - y_n}{x_{n+1} - x_n}$. Therefore $y_{n+1}  = \frac{dy}{dx}(x_n)h + y_n$.

### Example 1

```python
# Derivative
def deriv(y):
    return y

# Forward Euler
def fEuler(x, y, n, T):
    # Array of x and y values
    h = T / n
    x_n = [x]
    y_n = [y]
    x = x + h
    for i in range(0,n):
        x_n.append(x)
        y = deriv(y)*h + y
        y_n.append(y)
        x = x + h
    return x_n,y_n
```

```python
e = []
step = []

# Creates an array of the error between final y  value at various h values
for i in range(1,7):
    FE = fEuler(0,1,2**i,1)
    e.append(np.abs(FE[1][-1] - np.exp(1)))
    step.append(1 / 2**i)
```

```python
plt.loglog(step,e,label='error')
plt.xlabel('h')
plt.ylabel('error')

plt.loglog(step,step)

plt.legend()
```
