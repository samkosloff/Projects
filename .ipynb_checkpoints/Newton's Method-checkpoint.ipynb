## Objective
In this notebook I am attempting to create an algorithm which will find the roots of some function $f:\mathbb{R}\rightarrow\mathbb{R}$ given a guess of some point near where the zero might lie.

The general approach will be as follows:

1. Guess $g_1(g_1,y(g_1))$
2. The tangent line to that point is given by $y = f'(g_1)x + b$ where $b = f(g_1) - f'(g_1)g_1$. Therfore $y = f'(g_1) (x - g_1) + f(g_1)$
3. Find x-intercept: $x_1 = g_1 - \frac{f(g_1)}{f'(g_1)}$
4. Repeat from step (2) with next guess $g_2 = x_1$


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Polynomial #1

```python
# Define the coefficients of the polynomial equation
coefficients = [1,5,2,-30] # x^3 + 5x^2 + 2x - 30

# Create a polynomial object
f = np.poly1d(coefficients)
```

```python
# Generate x values for plotting
x = np.linspace(-5, 5, 50)

# Evaluate the polynomial for the corresponding x values
y = f(x)

# Plot the polynomial
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of the Polynomial')
plt.grid(True)

# Add a horizontal line at y = 0
plt.axhline(y=0, color='red', linestyle='-')

plt.show()
```

```python
g = 10
loops = 20
true_zero = 1.94001884310443

shots = []
for i in range(loops):
    b = f(g) - f.deriv()(g)*g
    g = g - ( f(g) / f.deriv()(g) )
    shots.append(g - true_zero)
```

```python
# Generate x and y values for plotting
x = np.linspace(0, 0.2, loops)

y = np.array(shots)

# Plot the polynomial
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Distance from True Zero')
plt.grid(True)

# Add a horizontal line at y = 0
plt.axhline(y=0, color='red', linestyle='-')

plt.show()
```
